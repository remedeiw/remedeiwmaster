```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datapl1 = pd.read_csv('PL1sum.csv')
datapl2 = pd.read_csv('Pl2sum.csv')
datapl3 = pd.read_csv('Pl3sum.csv')
# datalistclm = pd.read_csv(Data/Lastprofile/PL1.csv)
```

```python
data = datapl1
clmn = list(pd.read_csv('Data/Lastprofile/PL1.csv'))
for column in clmn:
    data[column] = datapl1[column]+datapl2[column]+datapl3[column]
data
```

```python
data["time"] = data["Tag"]
for i in range(0,35040):
    data["time"][i] = str(pd.Timestamp(year=2014, month=data["Monat"][i], day=data["Tag"][i], hour=data["Stunde"][i], minute=data["Minute"][i]))
data
```

```python
data = data.set_index('time')
data
```

```python
plotdatajan = data.loc['2014-02-01 00:00:00':'2014-02-28 23:00:00']
plt.figure(figsize = (12, 5))
plotdatajan['Profil 2'].plot()
plt.xticks(rotation=45)
plt.show()
```
